Tratamento de missings.

Os dados missings foram tratados com "-1", logo existem dados faltantes. A analise dirá qual tratamento a ser dado.

```python
from numpy.core.numeric import NaN 
df_train2 = df_train.replace(-1,NaN) # substituição do -1 por NaN para que possa utilizar as ferramentas da biblioteca "Missingno" no tratamento dos dados faltantes. 
```

```python
df_train2.info()
```

```python
df_train2.isnull().sum() # contagem do numero de dados faltantes e exibição das observações que possuem dados faltantes.
```

```python
missingValueColumns = df_train2.columns[df_train2.isnull().any()].tolist()
df_null = df_train2[missingValueColumns]
df_null # dataframe com as colunas que possuem dados faltantes (13 no total).
```

```python
df_null.isnull().sum()
```
